```python
import pandas as pd
import requests
```

```python
MRT_data = pd.read_csv('./DATA/臺北捷運每日分時各站OD流量統計資料.csv')

MRT_data
```

### Download Data

```python
# Create a directory to store downloaded files
# download_directory = r'C:\Python\NewYearResolution\DATA'
# os.makedirs(download_directory, exist_ok=True)

for index, row in MRT_data.iterrows():
    year_month = row['年月']
    url = row['URL']

    # Generate a filename based on '年月'
    filename = f"{year_month}.csv"

    # Combine the download directory and filename to get the full path
    # filepath = os.path.join(download_directory, filename)

    # Make a request to download the file
    response = requests.get(url)

    if response.status_code == 200:
        # Write the content to the file
        with open(f'DATA/MRT_raw_data/{filename}', 'wb') as file:
            file.write(response.content)
        print(f"File '{filename}' downloaded successfully.")
    else:
        print(f"Failed to download file from '{url}'. Status code: {response.status_code}")

# Now, all the files should be downloaded to the 'downloaded_files' directory.
```

### Merge Data

```python
import glob
import pandas as pd
```

```python

folder_path = "DATA/MRT_raw_data"
output_path = "DATA/MRT_data.csv"  # Change this to the desired output file name

# Use glob to get a list of all files in the folder matching the pattern
files = glob.glob(f"{folder_path}/*.csv")  # Adjust the file extension based on your data format

# Initialize an empty DataFrame to store the merged data
merged_data = pd.DataFrame()

# Loop through each file in the list
for file_path in files:
    # Read the data from the file into a DataFrame
    current_data = pd.read_csv(file_path)
    
    # Concatenate the current data to the merged_data DataFrame
    merged_data = pd.concat([merged_data, current_data], ignore_index=True)

# Write the merged data to a new CSV file
merged_data.to_csv(output_path, index=False)
```
